```python
df = pd.read_csv('whisper.csv')
for item in combined_list:
    df[item] = 0
df
```

```python
import re
sia = SentimentIntensityAnalyzer()
for index in df.index:
    scores = sia.polarity_scores(df.loc[index, 'text'])
    for i,item in enumerate(friends_and_family):
        val = 1
        if scores['compound']<=0.0 and re.search(re.escape(item),  df.loc[index, 'text'], re.IGNORECASE):
            val = 0
        df.loc[index, item] = val
```
